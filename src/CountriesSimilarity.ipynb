```python
from modules.spoticharts import SpotiCharts
from scipy.spatial import distance
from modules.plot_utils import plot_similarity
import numpy as np
import pandas as pd
```

```python
class SimilarityComparator:
    def __init__(self):
        self.sc = SpotiCharts()
        
    def __jaccard(self, a,b):
        return len(a & b) / len(a | b)
    
    def compare_similarity(self, first_country, second_country, start_date=None, end_date=None):
        a = self.sc.get_all_viral(country_codes=[first_country], start_date=start_date, end_date=end_date)
        b = self.sc.get_all_viral(country_codes=[second_country], start_date=start_date, end_date=end_date)
        dates = a['date'].values
        sorted_dates = sorted(list(set(dates)), reverse=True)
        x,y = [],[]
        for date in sorted_dates:
            sets = []
            for df in [a, b]:
                s = set(df.loc[df['date'] == date]['spotify_id'].values)
                sets.append(s)
            x.append(date)
            y.append(self.__jaccard(sets[0], sets[1]))
        return x, y, np.mean(y), np.median(y)
```

```python
comparator = SimilarityComparator()
```

```python
similarities_pl = {}
```

```python
def plot_countries_similarity(first_country, second_country, start_date=None, end_date=None):
    x,y, mean, median = comparator.compare_similarity(first_country, second_country, start_date, end_date)
    print("Average similarity:", mean)
    print("Median:", median)
    similarities_pl[second_country] = [mean, median]
    plot_similarity(x,y,'Jaccard similarity: {} {}'.format(first_country, second_country))
```
